```python
def translate_policy(row):
    if row['1j. On request (Yes/No/Law Varies By Jurisdiction)'] == 'Yes':
        return "On request"
    elif row['1i. For economic or social reasons (Yes/No/Law Varies By Jurisdiction)']
```

```python
def intepret_policy(policy_string):
    if policy_string == 'No,No,No,No':
        return 'No to all'
    elif policy_string == 'No,Yes,No,No' or policy_string == 'N/A,Yes,N/A,N/A':
        return 'Only to save women\'s life'
    elif policy_string == 'No,Yes,Yes,No':
        return 'To save women\'s life and health'
    elif policy_string == 'Yes,Yes,Yes,Yes':
        return 'Allow all' 
    else:
        return 'Others/Not specify'
```
